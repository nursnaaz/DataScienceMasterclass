```python
y = lambda x : 6*x**2 + 2*x + 10
```

```python
der = lambda x : 12*x + 2
```

```python
lr = 0.001
cur_x = (10)
change = 1
iteration = 100
```

```python
minimal = []
x_value = []
change__ = []
while abs(change) > 0.0001:
    #print(y(cur_x))
    minimal.append(y(cur_x))
    x_value.append(cur_x)
    prev_x = cur_x
    cur_x = cur_x - der(cur_x) * lr
    change = prev_x - cur_x
    #print(prev_x, cur_x, change)
    change__.append(change)
print(prev_x, cur_x, change)    
print(y(cur_x))
```

```python
minimal = []
x_value = []
change__ = []
while iteration > 0:
    #print(y(cur_x))
    minimal.append(y(cur_x))
    x_value.append(cur_x)
    prev_x = cur_x
    cur_x = cur_x - der(cur_x) * lr
    change = prev_x - cur_x
    #print(prev_x, cur_x, change)
    change__.append(change)
    iteration = iteration - 1
print(prev_x, cur_x, change)    
print(y(cur_x))
```
