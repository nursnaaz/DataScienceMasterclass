### Step 6. Capitalize both Mjob and Fjob

```python
df['Mjob'].apply(capt)
df['Fjob'].apply(capt)
```

```python
def capti(x):
    res = []
    for i in x:
        res.append(i.capitalize())
    return res
```

```python
df[['Mjob','Fjob']].apply(lambda x : capti(x))
```

### Step 9. Create a function called majority that return a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old)

```python
def majority(age):
    if age > 17:
        return True
    else:
        return False
```

```python
df['legal_drinker'] = df['age'].apply(majority)
```
